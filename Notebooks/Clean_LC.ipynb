# Filtering to CA, CO, OR, TX and WA only on: per_capita_alcohol_consumption_geographic_1977_2017_clean.csv¶

```python
csvpath4 = os.path.join('..', 'Data','Clean', 'per_capita_alcohol_consumption_geographic_1977_2017_clean.csv')
file_df4 = pd.read_csv(csvpath4)
```

```python
#Picking columns of Interest
reducing1_df = file_df4[['geographic area','Year','Wine']]

# # Filtering column "geographic area" to only contain: Colorado, California, Oregon, Texas and Washington
states_df = reducing1_df.loc[(reducing1_df['geographic area'] == 'Colorado') | (reducing1_df['geographic area'] == 'California') | (reducing1_df['geographic area'] == 'Oregon') | (reducing1_df['geographic area'] == 'Texas') |(reducing1_df['geographic area'] == 'Washington'), :]

#Generating 2009 only table
year2009_df = states_df.loc[(states_df['Year']==2009),:]

year2009_df = year2009_df.rename(columns={'Wine':'Wine2009'})

wine2009 = year2009_df[['Wine2009','geographic area']]

#Generating 2017 only table
year2017_df = states_df.loc[(states_df['Year']==2017),:]

year2017_df = year2017_df.rename(columns={'Wine':'Wine2017'})

wine2017 = year2017_df[['Wine2017','geographic area']]

#Merge 2017 and 2009 per capita wine consumption

merged_wine = pd.merge(wine2009,wine2017, how='outer', on='geographic area')
merged_wine
```

```python
# Generating New Data Frame Representing Wine consumption evolution from 2009-2017

merged_wine['Per Capita Increase(2009-2017)']=merged_wine['Wine2017']-merged_wine['Wine2009']
merged_wine
```

```python
output_file2 = os.path.join("..","Data","Clean","Per_CapitaWine_state_final.csv")

merged_wine.to_csv(output_file2, index=False, header=True)
```
