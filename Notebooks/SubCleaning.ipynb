```python
import pandas as pd
import os
```

```python
## Read in Accident CSV file
file = os.path.join("..","Cleaned", "2014_2018_merged.csv")
```

```python
data = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
data.columns
```

```python
df = data.drop(columns=['Unnamed: 0', 'ST_CASE', 'PERSONS', 'COUNTY', 'CITY', 'DAY','YEAR','MINUTE', 'ROUTE', 'TWAY_ID','TYP_INT',
    'TWAY_ID2','LATITUDE', 'LONGITUD','WEATHER1', 'WEATHER2', 'CF1', 'CF2', 'CF3', 'FATALS','PER_TYP', 'TRAV_SP','Year', 'ROAD_FNC'])
#                         ,'REL_JUNC','TRAF_FLO', 'NO_LANES', 'SP_LIMIT', 'ALIGNMNT', 'PROFILE',
#        'PAVE_TYP', 'SUR_COND', 'TRA_CONT', 'T_CONT_F', 'HIT_RUN', 'C_M_ZONE',
#        'latitude', 'longitud'])
```

```python
df.head()
```

```python
hours = []

for i in df['HOUR']:
    if i > 0 and i < 6:
        hours.append('Early Morning')
    elif i < 12:
        hours.append('Mid Morning')
    elif i < 18:
        hours.append('Mid Afternoon')
    else:
        hours.append('Party Time')

df['Time Group']= hours
df = df.drop(columns=['HOUR'])
df.head()
```

```python
inj = []

for i in df['INJ_SEV']:
    if i == 0:
        inj.append('0')
    elif i > 0 and i < 4:
        inj.append('1')
    elif i == 4:
        inj.append('2')
    else:
        inj.append('0')


df = df.drop(columns=['INJ_SEV'])
df['INJ_SEV']= inj
df.head()
```

```python
drugs = []

for i in df ['DRUGS']:
    if i == 0 or i >1:
        drugs.append(0)
    else:
        drugs.append(1)


df = df.drop(columns=['DRUGS'])
df['DRUG_USE'] = drugs
df.head()
```

```python
drinks = []

for i in df['DRINKING']:
    if i == 0 or i > 1:
        drinks.append(0)
    else:
        drinks.append(1)

df = df.drop(columns=['DRINKING'])
df['DRINKING'] = drinks
df.head()
```

```python
weather = []

for i in df['WEATHER']:
    if i == 8 or i > 10:
        weather.append(8)
    elif i == 9:
        weather.append(1)
    else:
        weather.append(i)
        
df = df.drop(columns=['WEATHER'])
df['WEATHER'] = weather
df.head()
```

```python
helmet = []

for i in df['REST_USE']:
    if i == 19 or i == 5 or i == 6 or i == 16 or i ==17:
        helmet.append(1)
    else:
        helmet.append(0)
        
df = df.drop(columns=['REST_USE'])
df['HELMET'] = helmet
df.head()
```

```python
prev_acc = []

for i in df['PREV_ACC']:
    if  i ==98 or i == 99 or i == 998:
        prev_acc.append(0)
    else:
        prev_acc.append(i)
        
df = df.drop(columns=['PREV_ACC'])
df['Previous Accident'] = prev_acc
df.head()
```

```python
prev_dwi = []

for i in df['PREV_DWI']:
    if  i ==98 or i == 99 or i == 998:
        prev_dwi.append(0)
    else:
        prev_dwi.append(i)
        
df = df.drop(columns=['PREV_DWI'])
df['Previous DWI'] = prev_dwi
df.head()
```

```python
prev_spd = []

for i in df['PREV_SPD']:
    if  i ==98 or i == 99 or i == 998:
        prev_spd.append(0)
    else:
        prev_spd.append(i)
        
df = df.drop(columns=['PREV_SPD'])
df['Previous Speeding Infractions'] = prev_spd
df.head()
```

```python
list(df.columns)
```

```python
cleaned = pd.get_dummies(df,columns=['MONTH', 'DAY_WEEK','SEX','DRUG_USE','DRINKING','WEATHER'])
# 'INJ_SEV',
```
