```python
import pandas as pd
import os
```

```python
## Read in Accident CSV file
```

```python
file = os.path.join("..","Raw_files", "FARS2018NationalCSV","ACCIDENT.csv")
```

```python
# create dataFrame and load in data from CSV file
```
